```python
import pandas as pd

data = pd.read_csv("Daily_Global_Stock_Market_Indicators.csv")
print(data)
```

```python
# change column names 
data.columns = data.columns.str.strip().str.lower().str.replace(' ', '_')
print(data.columns)
```

```python
# remove white spaces from all 
for i in data.select_dtypes(include=["object"]).columns:
    data[i] = data[i].astype(str).str.strip()

data
```

```python
data.info()
```

```python
# change date column data type object to date format
data['date'] = pd.to_datetime(data['date'])
data.info()

```

```python
# hanfling missing value
# if numeric valur then fill it with median and if categorical data then fill it with mode 

for col in data.columns:
    if pd.api.types.is_numeric_dtype(data[col]):
        data[col].fillna(data[col].median())
    elif pd.api.types.is_datetime64_any_dtype(data[col]):
        data[col].fillna(data[col].mode())
    elif pd.api.types.is_object_dtype(data[col]):
        data[col].fillna(data[col].mode())
    else:
        data[col].fillna(data[col].ffill())
        data[col].fillna(data[col].bfill())

data
```

```python
# remove duplicates
len(data)
data.drop_duplicates()
len(data)
```

```python
# add new column year, month, day
data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month
data['day'] = data['date'].dt.day
data
```

```python
# remove columns not required for analysis
data.drop(columns=["weekday"])
```

```python
# save cleaned data to new csv file
data.to_csv("CleanData_.csv")
```
