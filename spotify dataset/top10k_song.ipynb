```python
import pandas as pd
import numpy as np
import os

os.getcwd()
os.listdir()
df = pd.read_csv("top-10k-spotify-songs-2025-07.csv")
df.columns = df.columns.str.strip().str.lower().str.replace(" ","_") #strip for remove white space(column name)
df.columns
```

```python
print(df.info())
```

```python
 # remove white space in object data type column
for i in df.select_dtypes(include = ["object"]).columns:
  df[i] = df[i].astype(str).str.strip()

df
```

```python
print(df.isnull().sum())
```

```python
# normalize date time column
from datetime import date
#standarize date time columns
dates = [i for i in df.columns if any(j in i for j in["date","timestamp","time","dt"])]
dates
for cols in dates:
  try:
    df[cols] = pd.to_datetime(df["release_date"], errors='coerce')
  except:
    print(cols)

df.info()
```

```python
# create part of datetime columns 
for i in dates:
    if pd.api.types.is_datetime64_any_dtype(df[i]):
        base = i.rstrip("_")
        df[f"{base}_year"] = df[i].dt.year  
        df[f"{base}_month"] = df[i].dt.month  
        df[f"{base}_day"] = df[i].dt.day  

df
```

```python
# normalize N/A value

na_values = ["na","n/a","none","null",""] # represet as null value
for i in df.columns:
  df[i] = df[i].replace(na_values,np.nan)

df
```

```python
print(df.isnull().sum())
```

```python
# hanfling missing value
# if numeric valur then fill it with median and if categorical data then fill it with mode 

for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):
        df[col].fillna(df[col].median())
    elif pd.api.types.is_datetime64_any_dtype(df[col]):
        df[col].fillna(df[col].mode())
    else:
        df[col].fillna(df[col].ffill())

df
```

```python
# remove duplicates
len(df)
df.drop_duplicates()
len(df)
```

```python
df.to_csv("clean.csv")
```
