```python
import pandas as pd
import numpy as np
import os

os.getcwd()
df = pd.read_csv('Students Social Media Addiction.csv')
print(df.head(10))
```

```python
# clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ","_")
df.columns
```

```python
df.info()
```

```python
# remove white space from object data types values
for i in df.select_dtypes(include=["object"]).columns:
    df[i] = df[i].str.strip()

df
```

```python
# normalize N/A values

na_values = ["n/a", "na", "--", "none", "null", "nil", " ","nan"]

for i in df.columns:
    if df[i].dtype == "object":
        df[i] = df[i].replace(to_replace=na_values , value=np.nan)
df
```

```python
# handle missing value

for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):
        df[col] = df[col].fillna(df[col].median())
    elif pd.api.types.is_datetime64_any_dtype(df[col]):
       df[col] = df[col].fillna(df[col].mode())
    else:
       df[col] = df[col].fillna(df[col].ffill())
       df[col] =  df[col].fillna(df[col].bfill())

df
```

```python
# handle duplicates
len(df)
df.drop_duplicates()
len(df)
```

```python
df.to_csv("clean_data.csv")
```
